```python
admPriv = Sudeste[(Sudeste['Dependência Administrativa'] == 'Privada' ) & (Sudeste['Localização'] == 'Total')]
admPriv
```

Total de Aprovação e Reprovação das escolas privadas dos anos de 2016 à 2020

```python
ax = sns.displot(data=admPriv, x="Total_aprovacao", col="Ano", bins=10 )
plt.yscale('log')
ax = sns.displot(data=admPriv, x="Total.1_reprovacao", col="Ano", bins=10 )
plt.yscale('log')

```

Total de Aprovação e Reprovação das escolas públicas dos anos de 2016 à 2020

```python
ax = sns.displot(data=admPubli, x="Total_aprovacao", col="Ano", bins=10 )
plt.yscale('log')
ax = sns.displot(data=admPubli, x="Total.1_reprovacao", col="Ano", bins=10 )
plt.yscale('log')
```

Análise estatística dos índices de aprovação, reprovação e abandono de todas as escolas da região Sudeste

```python
Sudeste.groupby('Ano')[['Total_aprovacao', 'Total.1_reprovacao', 'Total.2_abandono']].agg(['mean', 'std', 'median'])
```

Análise estatística dos índices de aprovação, reprovação do 9º ano do ensino fundamental e da 3ª série do ensino médio, de todas as escolas da região Sudeste

```python
Sudeste.groupby('Ano')[['9º Ano_aprovacao', '9º Ano.1_reprovacao', '3ª série_aprovacao', '3ª série.1_reprovacao']].agg(['mean', 'std', 'median'])
```

Análise estatística das aprovações, reprovações e abandono das escolas privadas dos ultimos 5 anos

```python
admPriv.groupby('Ano')[['Total_aprovacao', 'Total.1_reprovacao', 'Total.2_abandono']].agg(['mean', 'std', 'median'])
```

Análise estatística das aprovações, reprovações e abandono das escolas públicas dos ultimos 5 anos

```python
admPubli.groupby('Ano')[['Total_aprovacao', 'Total.1_reprovacao', 'Total.2_abandono']].agg(['mean', 'std', 'median'])
```

Análise estatística das taxas de aprovação e reprovação das escolas públicas e privadas com distinção para localidade de rural e urbana e a junção entre as duas, dos últimos 5 anos

```python
pd.set_option('display.max_rows', 100)
Sudeste[Sudeste['Dependência Administrativa'].isin(['Privada', 'Pública'])].groupby(['Ano','Dependência Administrativa', 'Localização'])[['Total_aprovacao', 'Total.1_reprovacao']].agg(['mean', 'std', 'median'])
```

```python
pd.set_option('display.max_rows', 100)
Sudeste[(Sudeste['Dependência Administrativa'].isin(['Pública']))].groupby(['Ano','Dependência Administrativa', 'Localização'])[['Total_aprovacao', 'Total.1_reprovacao']].agg(['mean', 'std', 'median'])
```

```python
pd.set_option('display.max_rows', 100)
Sudeste[Sudeste['Dependência Administrativa'].isin(['Privada'])].groupby(['Ano','Dependência Administrativa', 'Localização'])[['Total_aprovacao', 'Total.1_reprovacao']].agg(['mean', 'std', 'median'])
```

Análise estatística de aprovação e reprovação das escolas publicas e privadas com foco nas sérias: 9º ano do ensino fundamental e 3º ano do ensino médio

```python
pd.set_option('display.max_rows', 100)
Sudeste[Sudeste['Dependência Administrativa'].isin(['Privada', 'Pública'])].groupby(['Ano','Dependência Administrativa', 'Localização'])[['9º Ano_aprovacao', '9º Ano.1_reprovacao', '3ª série_aprovacao', '3ª série.1_reprovacao']].agg(['mean', 'std', 'median'])
```

Análise estatística da taxa de abandono das escolas públicas e privadas dos últimos 5 anos

```python
pd.set_option('display.max_rows', 100)
Sudeste[Sudeste['Dependência Administrativa'].isin(['Privada', 'Pública'])].groupby(['Ano','Dependência Administrativa', 'Localização'])[['Total.2_abandono']].agg(['mean', 'std', 'median'])
```
